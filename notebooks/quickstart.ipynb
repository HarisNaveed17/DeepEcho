```python
from deepecho.benchmark import Simple1, Simple2

for benchmark in [Simple1(), Simple2()]:
    fig = benchmark.visualize()
    fig.suptitle(benchmark.__class__.__name__)
```

```python
from deepecho.model import AlphaModel

for benchmark in [Simple1(), Simple2()]:
    print("Benchmark:", benchmark.__class__.__name__)
    
    model = AlphaModel(use_sampling=False)
    results = benchmark.evaluate(model)
    print("  AlphaModel(use_sampling=False)\n    %s\n" % results)

    model = AlphaModel(use_sampling=True)
    results = benchmark.evaluate(model)
    print("  AlphaModel(use_sampling=True)\n    %s\n" % results)
```
